```python
import numpy as np
data=[1,1,1,2,3,4,12,6,8,9,11,13,202,789]
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
IQR=Q3-Q1
print(Q1)
print(IQR)
```

```python
LF=Q1-1.5*(IQR)
UF=Q3+1.5*(IQR)
print(LF)
print(UF)
```

```python
newData=[]
for i in data:
    if(i<=UF and i>=LF):
        newData.append(i)

print(newData)
```
